```python
data_dir = '../data/'
batch_size = 100
```

```python
CIFAR10_LABEL_NAMES = (
'airplane',
'automobile',
'bird',
'cat',
'deer',
'dog',
'frog',
'horse',
'ship',
'truck')
```

```python
## データの水増しと正規化
transform = transforms.Compose([
#     transforms.RandomRotation(20),
#     transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),
#     transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
#     transforms.Normalize(mean=[0.5, 0.5, 0.5],
#                         std=[0.2, 0.2, 0.2])
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

print('start to load train dataset')
trainset = datasets.CIFAR10(
    root=data_dir, train=True, download=True, transform=transform
)
trainloader = torch.utils.data.DataLoader(
    trainset, batch_size=batch_size, shuffle=True
)
```

```python
idx = 3
loader = iter(trainloader)
d = next(loader)
```

```python
def unnorm(tensor):
    mean=np.array([0.485, 0.456, 0.406])
    std=np.array([0.229, 0.224, 0.225])
    for t, m, s in zip(tensor, mean, std):
        t.mul_(s).add_(m)
    return tensor
```

```python
fig, axs = plt.subplots(5, 8, figsize=(25, 15))
for i in range(5):
    for j in range(8):
        idx = 8*i + j
        img = d[0][idx]
        label = d[1][idx]
        img = unnorm(img)
        img_numpy = img.numpy()
        img_numpy = img_numpy.transpose(1,2,0)
#         img_numpy = img_numpy * std + mean
        axs[i,j].imshow(img_numpy)
        axs[i,j].set_title('{}'.format(CIFAR10_LABEL_NAMES[label]))
        axs[i,j].set_xticklabels([]) 
        axs[i,j].set_yticklabels([]) 
                
```

## Cifar10の画像のcsvを作る

```python
import glob
import os
import pandas as pd
```

```python
train_img = glob.glob('../data/cifar10_pictures/train/*')
test_img = glob.glob('../data/cifar10_pictures/test/*')
```

```python
train_label = [os.path.basename(f).split('_')[0] for f in train_img]
test_label = [os.path.basename(f).split('_')[0] for f in test_img]
```

```python
df_train = pd.DataFrame()
df_train['img'] = [os.path.basename(i) for i in train_img]
df_train['label'] = train_label
df_train['label_index'] = df_train['label'].map(lambda x: CIFAR10_LABEL_NAMES.index(x))
df_test = pd.DataFrame()
df_test['img'] = [os.path.basename(i) for i in test_img]
df_test['label'] = test_label
df_test['label_index'] = df_test['label'].map(lambda x: CIFAR10_LABEL_NAMES.index(x))
```

```python
train_num = 100
test_num = 30
seed = 123
df_train_pick = pd.DataFrame()
df_test_pick = pd.DataFrame()
for label in CIFAR10_LABEL_NAMES:
    df_train_pick = df_train_pick.append(df_train[df_train.label == label].sample(n=train_num, random_state=seed))
    df_test_pick = df_test_pick.append(df_test[df_test.label == label].sample(n=test_num, random_state=seed))
        
```

```python
df_train_pick = df_train_pick.sort_index()
df_test_pick = df_test_pick.sort_index()
```

```python
df_train_pick.to_csv('../data/cifar10_pictures/train_small.csv', index=None)
df_test_pick.to_csv('../data/cifar10_pictures/test_small.csv', index=None)
df_train.to_csv('../data/cifar10_pictures/train_big.csv', index=None)
df_test.to_csv('../data/cifar10_pictures/test_big.csv', index=None)
```

## 専用のデータ・セット作成

```python
class Cifar10DataSet(Dataset):
    """Face Landmarks dataset."""

    def __init__(self, csv_file, root_dir, transform=None):

        self.image_dataframe = pd.read_csv(csv_file)
        self.root_dir = root_dir
        self.transform = transform

    def __len__(self):
        return len(self.image_dataframe)

    def __getitem__(self, idx):
        img_name = os.path.join(self.root_dir,
                                self.image_dataframe.loc[idx, 'img'])
#         image = io.imread(img_name)
        image = Image.open(img_name)
        label = self.image_dataframe.loc[idx, 'label_index']

        if self.transform:
            image = self.transform(image)

        return image, label
```

```python
csv_file = '../data/cifar10_pictures/train_small.csv'
root_train = '../data/cifar10_pictures/train/'
root_test = '../data/cifar10_pictures/test/'
```

```python
df = pd.read_csv(csv_file)
im_name = os.path.join(root_train, df.loc[0, 'img'])
im = io.imread(im_name)
```

```python
## データの水増しと正規化
transform = transforms.Compose([
    transforms.RandomRotation(20),
    transforms.RandomAffine(degrees=0, translate=(0.2, 0.2)),
    transforms.RandomHorizontalFlip(),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406],
                        std=[0.229, 0.224, 0.225])
])

dataset = Cifar10DataSet(csv_file, root_train, transform)
trainloader = torch.utils.data.DataLoader(
    dataset, batch_size=batch_size, shuffle=True
)
```
